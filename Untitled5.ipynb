```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the URL of the webpage containing Tesla revenue data
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"

# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the webpage
    soup = BeautifulSoup(response.text, "html.parser")

    # Find the table containing revenue data
    revenue_table = soup.find("table", {"class": "table"}).find("tbody")

    # Initialize lists to store data
    dates = []
    revenues = []

    # Extract data from the table
    for row in revenue_table.find_all("tr"):
        columns = row.find_all("td")
        if len(columns) == 2:
            date = columns[0].text.strip()
            revenue = columns[1].text.strip()
            dates.append(date)
            revenues.append(revenue)

    # Create a DataFrame
    tesla_revenue = pd.DataFrame({"Date": dates, "Revenue": revenues})

    # Display the last five rows using the tail function
    print(tesla_revenue.tail())
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Function to retrieve stock data and create a graph with a title
def make_graph(ticker_symbol, title):
    # Use yfinance to download stock data
    stock_data = yf.download(ticker_symbol)

    # Plot the stock data
    plt.figure(figsize=(12, 6))  # Set the figure size
    plt.plot(stock_data['Adj Close'], label=ticker_symbol)  # Adjusted Close price
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.title(title)  # Set the title
    plt.legend()  # Show the legend

    # Display the graph
    plt.show()

# Define the ticker symbol for Tesla (TSLA) and the title
ticker_symbol = "TSLA"
graph_title = "Tesla Stock Price"

# Call the make_graph function to create and display the graph
make_graph(ticker_symbol, graph_title)
```
